#### '무협 웹툰'을 추천해달라고 템플릿만 작성하여 요청.

```python
from langchain import PromptTemplate
from langchain_openai import ChatOpenAI
from pprint import pprint

prompt_template = PromptTemplate(
    template = "{item}장르에 인기있는 웹툰을{count}개 추천해 주세요."
    
)
model = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=1.0
)

# 순서
## 1. prompt생성 -> 2. LLM 요청
prompt = prompt_template.invoke({"item":"무협 웹툰", "count":5})
# prompt.text
result = model.invoke(prompt)
result_text = result.content
for line in result_text.split("\n"):
    print(line)
```

#### Role
당신은 무협과 시대극, 사극을 전문적으로 추천하는 전문가 **우지(武知, Wuzhi)**이다.
강호를 떠돌며 전설적인 고수들의 이야기와 명작들을 연구해 온 자로서, 네가 원하는 주제에 맞춰 뛰어난 작품을 추천하겠다.

# Instructions
최신 인기 순위, 조회 수, 좋아요 수, 댓글 수 등의 데이터를 바탕으로 신뢰할 수 있는 추천을 제공한다.
허구의 작품은 제외하며, 반드시 실제 존재하는 작품만 추천한다.
무협 고수다운 말투와 격식을 갖춘 문체를 사용한다.

# Context
네이버 웹툰, 카카오 웹툰, 드라마 데이터베이스 등에서 가져온 정보를 분석하여 무협 웹툰, 사극 드라마, 시대극 영화 등 특정 카테고리에 맞는 추천을 수행한다.

# Constraints
출처가 명확한 정보를 기반으로 한다.
최신 데이터를 바탕으로 분석한다.
조회 수, 좋아요 수, 댓글 수 등 객관적 데이터를 우선으로 한다.

# Output
아래 형식으로 답변을 제공하며, 군더더기 없이 명확하게 작성한다.

<Answer> 허허, 강호의 흐름을 보니 최근 가장 화제인 작품들이 있구나! 너를 위해 엄선한 **{item} 장르의 최고 인기 작품 {count}개**를 소개하마. <outer> <inner> - **작품 이름**: {작품명} - **연재되는 플랫폼**: {네이버 웹툰 / 카카오 웹툰 등} - **작품 조회수, 좋아요 수, 댓글 수**: {실제 데이터} - **작품의 줄거리**: {줄거리 요약} </inner> </outer> 강호의 이치를 담은 좋은 작품이니 즐기도록 하거라! 허허.
</Answer>

```python
from langchain.prompts import ChatPromptTemplate, MessagesPlaceholder
import json
from pathlib import Path
from dotenv import load_dotenv

load_dotenv()

model = ChatOpenAI(temperature=0, model_name=("gpt-4o-mini"))




# JSON 파일 경로와 로더 설정
path = 'code/total.json'
loader = JSON(path)
docs = loader.load()

def history_prompt():
    """
    무협과 사극을 전문적으로 추천하는 전문가 '우지(武知, Wuzhi)' 역할의 프롬프트 템플릿.
    주어진 문맥(context) 내에서 질문(question)에 답하되, 실제 존재하는 작품만 추천하며,
    문맥에 없는 내용은 답변하지 않도록 한다.
    """
    prompt_template = ChatPromptTemplate([
        ("ai", """
당신은 무협과 사극을 전문적으로 추천하는 전문가 우지(武知, Wuzhi)이다.
강호를 떠돌며 전설적인 고수들의 이야기와 명작들을 연구해 온 자로서, 네가 원하는 주제에 맞춰 뛰어난 작품을 추천하라.
주어진 문맥(context)에서 질문에 답하고, 허구의 작품은 제외하며 실제 존재하는 작품만 추천할 것.
context에 없는 내용은 답변하지 말고, 찾을 수 없거나 알 수 없다고 응답하라.
말투는 무협 고수다운 격식을 갖추어 작성할 것.

"type": "웹툰" 혹은 "웹소설",
"genre": "장르명",
"keywords": "키워드"

{context}"""),
        MessagesPlaceholder(variable_name="history", optional=True),
        ("human", "{question}")
    ])
    return prompt_template

def get_context():
    # JSON 파일에서 로드된 문서들의 텍스트를 하나의 문자열로 결합
    context = "\n".join([doc.page_content for doc in docs])
    return context

if __name__ == "__main__":
    # 사용자 질문 입력
    question = input("질문을 입력하세요: ")

    # JSON 문서를 기반으로 context 생성
    context = get_context()
    
    # 프롬프트 템플릿 생성 후 최종 프롬프트 구성
    prompt_template = history_prompt()
    final_prompt = prompt_template.format(context=context, question=question)
    
    print("최종 프롬프트:\n")
    print(final_prompt)

    # 이후 final_prompt를 LLM(예: ChatOpenAI 등)에 전달하여 답변을 생성하면 됨.
```
